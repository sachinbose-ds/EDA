```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from dash import Dash, html, dcc
import plotly.graph_objects as go
```

```python
# Load Dataset
url = 'https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv'
df = pd.read_csv(url)
```

```python
# Initial Exploration
df.head()
```

```python
#  Data Info and Summary
df.info()
df.describe()
```

```python
#  Data Cleaning
df.fillna(0, inplace=True)
df['date'] = pd.to_datetime(df['date'])
```

```python
# Summary Statistics
df.describe()
```

### Total Cases Over Time for a Specific Country

```python

country = 'India'
df_country = df[df['location'] == country]
fig1 = px.line(df_country, x='date', y='total_cases', title=f'Total Cases Over Time in {country}')
fig1.show()
```

```python
country = 'United States'
df_country = df[df['location'] == country]
fig1 = px.line(df_country, x='date', y='total_cases', title=f'Total Cases Over Time in {country}')
fig1.show()
```

### New Cases Over Time for a Specific Country

```python
country = 'India'
fig2 = px.line(df_country, x='date', y='new_cases', title=f'New Cases Over Time in {country}')
fig2.show()
```

### Total Deaths Over Time for a Specific Country

```python

fig3 = px.line(df_country, x='date', y='total_deaths', title=f'Total Deaths Over Time in {country}')
fig3.show()
```

### New Deaths Over Time for a Specific Country

```python

fig4 = px.line(df_country, x='date', y='new_deaths', title=f'New Deaths Over Time in {country}')
fig4.show()
```

### Total Cases Comparison Between Countries

```python

countries = ['United States', 'India', 'Brazil', 'Russia', 'United Kingdom']
df_countries = df[df['location'].isin(countries)]
fig5 = px.line(df_countries, x='date', y='total_cases', color='location', title='Total Cases Comparison')
fig5.show()
```

#  Dashboard Setup

```python

app = Dash(__name__)

app.layout = html.Div([
    html.H1('COVID-19 Data Dashboard'),
    dcc.Graph(figure=fig1),
    dcc.Graph(figure=fig2),
    dcc.Graph(figure=fig3),
    dcc.Graph(figure=fig4),
    dcc.Graph(figure=fig5),
    dcc.Graph(figure=fig6),
    dcc.Graph(figure=fig7),
    dcc.Graph(figure=fig8),
    dcc.Graph(figure=fig9)
])

if __name__ == '__main__':
    app.run_server(debug=True)
```
